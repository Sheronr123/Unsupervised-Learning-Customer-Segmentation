```python
# Now that we have done basic EDA, lets deal with kmeans algorithm:
# Before we do that lets scale our data:
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
scaled_df = ss.fit_transform(df)
```

```python
scaled_df
```

```python
from sklearn.cluster import KMeans
```

```python
# Firstly we want to find the best number of clusters, which we will use the elbow method:

inertia = []
for k in range(1, 15):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_df)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(1, 15), inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()

```

```python
# As we can see from the graph, the number of ideal clusters is 6-7, lets go with 7!
```

```python
kmeans = KMeans(n_clusters=7, random_state=42)
```

```python
kmeans.fit(scaled_df)
label = kmeans.labels_
label
```

```python
kmeans.cluster_centers_.shape
```

```python
# Lets obtain the cluster centres:
kmeans.cluster_centers_
```

```python
# Lets add this to a dataframe:
cluster_centres = pd.DataFrame(kmeans.cluster_centers_, columns=df.columns)
cluster_centres
```

```python
# Since the above are scaled data, there isnt much information we can extract, we have to transform back to our
# original df:
```

```python
# Lets transform this:
cluster_centres = ss.inverse_transform(cluster_centres)
cluster_centres = pd.DataFrame(cluster_centres, columns=df.columns)
cluster_centres
```

```python
import matplotlib.pyplot as plt

def plot_cluster_histograms(df, n_clusters):
    """
    Plots histograms for each column in the dataset, grouped by the 'Cluster' column.
    
    Parameters:
    df (pd.DataFrame): The dataset containing the 'Cluster' column.
    n_clusters (int): The number of clusters.
    """
    columns = df.columns.drop('Cluster')  # Exclude the 'Cluster' column
    n_columns = len(columns)
    
    for column in columns:
        plt.figure(figsize=(20, 5))
        plt.suptitle(f'Histograms of {column} by Cluster', fontsize=16)
        
        for i in range(n_clusters):
            plt.subplot(1, n_clusters, i + 1)
            cluster_data = df[df['Cluster'] == i][column]
            plt.hist(cluster_data, bins=15, edgecolor='black', alpha=0.7)
            plt.title(f'Cluster {i}')
            plt.xlabel(column)
            plt.ylabel('Frequency')
        
        plt.tight_layout()
        plt.show()

# Example usage
n_clusters = df['Cluster'].nunique()  # Get the number of unique clusters
plot_cluster_histograms(df, n_clusters)

# This below figures shows us the spread between each column for each cluster:
```
